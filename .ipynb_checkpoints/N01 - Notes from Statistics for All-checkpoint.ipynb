### Relative Frequency

**$fr_i = \frac{f_i}{\displaystyle \sum_{i=1}^{n} f_i}$**

````python
x  = [ 1,  2,  3,  4,  5]
y  = [43, 15, 22, 16,  4]

ty = sum(y)
fry = [elem / ty for elem in y]

plt.xticks([1, 2, 3, 4, 5], ['1-Priv', '2-GOVF', '3-GOVE/M', '4-NP', '5-NDA'])

plt.bar(x, fry, label = "Frequência R", color='r')

plt.xlabel("Tipo de ocupação")
plt.ylabel("Número de alunos")
plt.title("Ocupação x Frequência Relativa")

plt.plot(x, fry, label="Polígono", color='k')

plt.legend()
plt.show()
````

### Distribuição Hipergeométrica

Uma empresa remeteu 25 peças para uma loja. Mas 5 destas tinham defeito Quando foi descoberto, 4 das 25 tinham sido vendidas. Qual a probabilidade de 3 das 4 serem perfeitas e uma ter defeito?

<img src="images/probhip.jpg"/>

````python
# Definindo como sucesso encontrar defeitos.
N = 25 # Tamanho da pop.
r = 5  # Número de defeitos (sucesso na pop.)
n = 4  # Tamanho da amostra
x = 1  # sucessos ou defeitos na amosta

crx = factorial(r)/(factorial(r-x)*factorial(x))

cNrnx = factorial(N-r)/(factorial(N-r-n+x)*factorial(n-x))

cNn = factorial(N)/(factorial(N-n)*factorial(n))

px = (crx*cNrnx)/cNn
px
````

Funciona pensando de maneira inversa:

````python
# Definindo como sucesso encontrar perfeitos.
N = 25 # Tamanho da pop.
r = 20 # Número de perfeitos (sucesso na pop.)
n = 4  # Tamanho da amostra
x = 3  # sucessos ou perfeitos na amosta

crx = factorial(r)/(factorial(r-x)*factorial(x))
cNrnx = factorial(N-r)/(factorial(N-r-n+x)*factorial(n-x))
cNn = factorial(N)/(factorial(N-n)*factorial(n))

px = (crx*cNrnx)/cNn
px
````

### Distribuição de Poisson

Uma empresa oferece um exame gratuito de seus produtos por 7 dias. Foi observado que 2 em cada 10 produtos vendidos são devolvidos. Usando Poisson, encontrar a probabilidade de exatamente 6 produtos de 40 vendidos serem devolvidos.

<img src="images/probpoi.jpg"/>

2 em 10 produtos são devovidos. Para 40 produtos, lambda = 8 (lb)

````python
lb = 8
x  = 6

px6 = ((lb**x)*(np.exp(1)**(-lb)))/factorial(x)
px6
````

## Distribuição Normal

**Obtendo o valor de probabilidade para um valor de $z$ da distribuição Normal Padrão**

Seja uma distribuição Normal padrão $N(0,1)$ com média $\mu = 0$ e desvio-padrão $\sigma = 1$. Para obter o valor da probabilidade acumulada correspondente a uma faixa de $-\infty$ até o valor de $z$, isto é $P(-\infty \leq z \leq 0)$, deve-se empregar o comando **norm.cdf**.

````python
# Função de Distribuição Acumulada

mu    = 0                    # média igua a zero 
sigma = 1                    # desvio padrão igual a um
z     = 0.0                  # De menos infinito até zero

P1 = norm.cdf(z, mu, sigma)  # Probabilidade acumuada de menos infinito até o zero, qual é a probabiidade
P1
````

**Cálculo da probabilidade $P(0 \leq z \leq 1.95)$**

A partir do comando **norm.cdf**, que fornece a probabilidade acumulada $P(-\infty \leq z \leq 1.95)$, pode-se obter a probabilidade $P(0 \leq z \leq 1.95)$ utilizando-se $P(-\infty \leq z \leq 1.95)$ - $P(-\infty \leq z \leq 0.00)$.

````python
# Área entre zero e z=1.95

z     = 1.95

P2 = norm.cdf(z, mu, sigma)

P3 = P2 - P1  # Entre áreas

print("P(-inf<= z <= 0.00) = ",P1)
print("P(-inf<= z <= 1.95) = ",P2)
print("P(0.00<= z <= 1.95) = ",P3)
````

**Obtendo o valor de $P(-2.17 \leq z \leq 0.00)$**

Encontrar a área sob a curva da normal padronizada entre $z =
-2,17$ e $z = 0.00$. Para tanto, será necessário calcular:
$P(-\infty \leq z \leq 0.00)$ - $P(-\infty \leq z \leq -2.17)$.

````python
mu = 0
sigma = 1

z1 = -2.17
z2 =  0.00

# Acumulada entre -infinito até z1: -2.17
P1 = norm.cdf(z1, mu, sigma)

# Acumulada entre -infinito até z2:  0.00
P2 = norm.cdf(z2, mu, sigma)

print("P1 = P(-inf <= z <= -2.17) = " + str(P1))
print("P2 = P(-inf <= z <=  0.00) = " + str(P2))
print("P(-2.17 <= z <= 0.00) = " + str(P2-P1))
````

**Calcular a probabilidade de P(z > 2.32)**

Observar que o comando norm.cdf permite o cálculo da probabilidade  P(−∞≤z≤z1) , onde  z1=2.32 . Então, o cálculo de  P(z≥2.32)  pode ser realizado por meio de  1−P(−∞≤z≤2.32) , pois  P(−∞≤z≤2.32)+P(z≥2.32)=1

````python
mu = 0
sigma = 1
z1 = 2.32

# Acumulada entre -infinito até z1: 2.32
P = norm.cdf(z1, mu, sigma)

print("P(-inf <= z <= 2.32) = " + str(P))
print("P(z >= 2.32) = 1 - P(-inf <= z <= 2.32) = " + str(1-P))
````

**Calcular a probabilidade  P(−1,56 ≤ z ≤ 2,31)**

````python
mu = 0
sigma = 1

z1 = -1.56
z2 =  2.31

# Acumulada entre -infinito até z1: -2.17
P1 = norm.cdf(z1, mu, sigma)

# Acumulada entre -infinito até z2:  0.00
P2 = norm.cdf(z2, mu, sigma)

print("P1 = P(-inf <= z <= -1.56) = " + str(P1))
print("P2 = P(-inf <= z <=  2.31) = " + str(P2))
print("P(-1.56 <= z <= 2.31) = " + str(P2-P1))
````

**Calcular a probabilidade $P(z \geq -0,75)$**

Observar que o comando **norm.cdf** permite o cálculo da probabilidade $P(-\infty \leq z \leq z_1)$, onde $z_1 = -0.75$. Então, o cálculo de $P(z \geq -0.75)$ pode ser realizado por meio de $1 - P(-\infty \leq z \leq -0.75)$, pois $P(-\infty \leq z \leq -0.75) + P(z \geq -0.75) = 1$.

````python
mu = 0
sigma = 1

z1 = -0.75

# Acumulada entre -infinito até z1: 2.32
P = norm.cdf(z1, mu, sigma)

print("P(-inf <= z <= 0.75) = " + str(P))
print("P(z >= 0.75) = 1 - P(-inf <= z <= 0.75) = " + str(1-P))
````

**Calcular a probabilidade  P(25 ≤ x ≤ 32.8), sendo a distribuição fora do padrão (mu=25, sigma=4)**

````python
# norm.cdf faz a conversão automática quando o desvio padrão é diferente de 1

#z = (x-mu)/sigma

mu = 25
sigma = 4

x1 = 25    # z = (25-25)/4 = 0
x2 = 32.8

# Acumulada entre -infinito até x1: 25
P1 = norm.cdf(x1, mu, sigma)

# Acumulada entre -infinito até x2: 32.8
P2 = norm.cdf(x2, mu, sigma)

print("P1 = P(-inf <= x <= 25) = " + str(P1))
print("P2 = P(-inf <= x <=  32.8) = " + str(P2))
print("P(25 <= z <= 32.8) = " + str(P2-P1))
````

**Exemplo do Porsche 911**

Um guia automotivo indica o Porsche 911 como o carro que
melhor mantém valor. É esperado que um carro novo de R\\$
87.500 mantenha um valor de R\\$ 48.125 após 3 anos de uso.
Supondo que a distribuição de preços de todos os Porsches 911
com 3 anos seja normal com média R\\$ 48.125 e desvio-padrão de
R\\$ 1.600, encontrar a probabilidade de que 1 veículo
aleatoriamente escolhido tenha preço de venda entre R\\$ 46.000
e R\\$ 49.000.

````python
mu = 48125
sigma = 1600

x1 = 46000
x2 =  49000

# Acumulada entre -infinito até x1: 46000
P1 = norm.cdf(x1, mu, sigma)

# Acumulada entre -infinito até x2: 49000
P2 = norm.cdf(x2, mu, sigma)

print("P1 = P(-inf <= x <= 46000) = " + str(P1))
print("P2 = P(-inf <= x <=  49000) = " + str(P2))
print("P(46000 <= x <= 49000) = " + str(P2-P1))
````

**Exemplo da Calculadora**

A vida útil de uma calculadora possui uma distribuição normal com ma de 54 meses e d-p de 8 meses. A empresa garante que qualquer calculadora que comece a apresentar defeitos, dentro do período de 36 meses após a compra, será substituída por 1 nova. Qual a percentagem aproximada de calculadoras serão substituídas?

Resolução:

Seja  x  a vida útil da calculadora. Seja  x  normal com  μ=54  e  σ=8 . Para se calcular o valor de  P(x ≤ 36) , usa-se:

````python
mu = 54
sigma = 8

x = 36

# Acumulada entre -infinito até x: 36
P = norm.cdf(x, mu, sigma)

print("Porcentagem = 100 * P(-inf <= x <= 36) = " + str(P*100))
````

**Obtendo o valor de  z  para uma probabilidade  P  com distribuição Normal**

Seja uma distribuição Normal padrão  N(0,1)  com média  μ=0  e desvio-padrão  σ=1 . Para obter o valor de  z  associado à probabilidade acumulada correspondente a uma faixa de  −∞  até o valor de  z , isto é  P(−∞ ≤ z ≤ 0) , deve-se empregar o comando *norm.ppf*.

````python
mu = 0
sigma = 1

p = 0.50

z = norm.ppf(p,mu,sigma)
z
````

````python
# Para uma curva fora do padrão normal

mu = 25
sigma = 4

p = 0.50

x = norm.ppf(p,mu,sigma)
x
````

**Exemplo da calculadora: dado  P=0.01  achar  z  correspondente**

Utilizando os dados do Exemplo da Calculadora, determinar o período de garantia de modo que não mais de 1% das calculadoras vendidas sejam substituídas.

````python
mu = 54
sigma = 8

p = 0.01

x = norm.ppf(p,mu,sigma)

print("A garantia deverá ter",round(x,2)," meses")
````

**Mas, e se a garantia tivesse que ser um número inteiro?**

````python
mu = 54
sigma = 8

x = 35

# Acumulada entre -infinito até x: 35
P = norm.cdf(x, mu, sigma)

print("Porcentagem = 100*P(-inf <= x <= 35) = ", round(P*100,2))
````

**Aproximando a Binominal pela Normal**

Quando n é grande a aplicação da fórmula da Binomial é complicada e a distribuição normal pode ser usada como aproximação para a binomial desde que  np>5  e  nq>5 . Seja  n=30 ,  p=0.50 . Deseja-se calcular o valor da probabilidade  P(x=19) . Sabendo-se que a média  μ  da Binomial é obtida com  μ=n∗p  e  σ=(n⋅p⋅q)1/2 . Em termos de distribuição normal, a probabilidade equivalente que deverá ser calculada é  P(18.5≤ z ≤19.5) .

````python
n = 30
p = 0.50
q = 1 - p
x = 19  
mu = n*p
sigma = math.sqrt(n*p*q)

x1 = 18.5
x2 = 19.5

# Acumulada entre -infinito até x1: 18.5
P1 = norm.cdf(x1, mu, sigma)

# Acumulada entre -infinito até x2: 19.5
P2 = norm.cdf(x2, mu, sigma)

print("P1 = P(-inf <= x <= 18.5) = ", round(P1,4))
print("P2 = P(-inf <= x <= 19.5) = ", round(P2,4))
print("P(18.5 <= x <= 19.5) = ", round(P2-P1,4))
````
